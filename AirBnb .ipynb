## CONCLUSION

* Manhattan is the undisputed luxury leader with the highest average price (~208 Dollar) and highest cost per bed (140 Dollor).
* While the average price is 187 Dollar, the Statistical Summary revealed extreme outliers up to 100,000 Dollar.
* For travelers looking for value, the Bronx and Staten Island are the best choices, offering significantly lower average prices (~$108-$118).
* Feature Engineering, we found that Staten Island offers the lowest "Price per Bed" (~$67), making it the most cost-effective borough for groups.
* The availability_365 graph shows a massive spike at the 350-365 day mark. This proves that NYC Airbnb is dominated by professional, full-time hosts rather than casual "home-sharers."
* We observed a slight negative correlation between price and reviews, suggesting that lower-priced units attract higher guest turnover and more frequent feedback.
